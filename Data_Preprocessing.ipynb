```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# Problem 1
# Inputs: a GenePattern GCT file, An optional summary parameter, boolean, with a default value of False.
# Output: a Pandas DataFrame containing the contents of the input file

def read_gct(filepath, summary=False):
    df = pd.read_csv(filepath, sep='\t', skiprows=2)
    df.set_index(['Name', 'Description'], inplace=True)
    if summary:
        print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    return df
```

```python
# Problem 2A

min_df = read_gct("BRCA_minimal_60x19.gct", True)
min_df
```

```python
# Problem 2B

df = read_gct("BRCA_large_20783x40.gct")
df
```

```python
# Problem 3A
# Using the data in the DataFrame from #2b, plot a histogram for each of the following columns. 
# Limit the maximum histogram value to 50000. Make sure that your histograms are labeled with the
# sample name that they are displaying (A7-A0DB-normal)

samples = ['A7-A0DB-normal', 'A7-A13E-normal', 'BH-A0B3-primary', 'BH-A0B5-primary']

for sample in samples:
    data = df[sample]
    plt.figure(figsize=(6, 4))
    plt.hist(data, bins=50, range=(0,50000))
    plt.title(sample)
    plt.xlim(0, 50000)
    plt.xlabel('Expression Level')
    plt.ylabel('Frequency')
    plt.tight_layout()
```

```python
# Problem 4
# Using the data in the DataFrame from #2b, print out a DataFrame containing one row for each 
# sample, with the following columns computed: Sample name, Mean, Median, Standard deviation

samples = df.columns 

stats = pd.DataFrame(columns=['Sample Name', 'Mean', 'Median', 'Standard Deviation'])
for i, sample in enumerate(samples):
    data = df[sample]
    mean = data.mean()
    med = data.median()
    sd = data.std()
    stats.loc[i] = [sample, mean, med, sd]
print(stats)
```

```python
# Problem 5A
# Filter out (remove) the rows whose maximum value is less than 1000.

row_max = df[samples].max(axis=1)
df_filtered = df[row_max >= 1000]
df_filtered
```

```python
# Problem 5B
# What are the mean and median of each column after applying this filter? 
# Plot the histograms of the same 4 columns from question #3.

samples = ['A7-A0DB-normal', 'A7-A13E-normal', 'BH-A0B3-primary', 'BH-A0B5-primary']
stats_filtered = pd.DataFrame(columns=['Sample Name', 'Mean', 'Median'])
for i, sample in enumerate(samples):
    if stats.loc[i]['Sample Name'] == sample:
        mean = stats.loc[i]['Mean']
        median = stats.loc[i]['Median']
        stats_filtered.loc[i] = [sample, mean, median]
    
for sample in samples:
    plt.hist(df_filtered[sample], bins=50, range=(0,50000))
    plt.title(sample)
    plt.xlim(0, 50000)
    plt.xlabel('Expression Level')
    plt.ylabel('Frequency')
    plt.show()
```
